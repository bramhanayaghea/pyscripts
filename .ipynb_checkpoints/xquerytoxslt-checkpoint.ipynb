```python
import lxml.etree as etree

def xquery_to_xslt(xquery_str, output_xslt_file, root_element_name=None):
    """
    Converts an XQuery string to an XSLT string and saves it to a file.

    Args:
        xquery_str (str): The XQuery code as a string.
        output_xslt_file (str): The path to the output XSLT file.
        root_element_name (str): The root element name to wrap the resulting XSLT code.
    """
    print(xquery_code)
    # If a root element name is provided, wrap the XQuery code with it.
    if root_element_name:
        xquery_str = f"<{root_element_name}>\n{xquery_str}\n</{root_element_name}>"

    # Parse the XQuery string.
    xquery_tree = etree.fromstring(xquery_str)

    # Create an empty XSLT stylesheet.
    xslt_tree = etree.ElementTree()
    xslt_stylesheet = etree.Element('xsl:stylesheet', version='1.0', xmlns='http://www.w3.org/1999/XSL/Transform')
    xslt_tree.setroot(xslt_stylesheet)

    # Iterate over the XQuery expressions in the file and convert them to XSLT templates.
    for xqy_expression in xquery_tree.iterfind('.//xquery:expr'):
        xslt_template = etree.Element('xsl:template', match='*')
        xslt_template.append(etree.fromstring(xqy_expression.text))
        xslt_stylesheet.append(xslt_template)

    # Return the XSLT stylesheet as a string.
    xslt_str = etree.tostring(xslt_tree, encoding='utf-8').decode('utf-8')

    # Write the XSLT file to the specified path.
    with open(output_xslt_file, 'w', encoding='utf-8') as f:
        f.write(xslt_str)

# Example usage:
xquery_code = """
declare namespace ns2="http://adpolice.gov.ae/";
declare namespace ns1="http://microservices.tamm.abudhabi.ae/ticketservice";
declare variable $requestElement as element() (:: schema-element(ns1:getMaterials) ::) external;

declare function local:func($requestElement as element() (:: schema-element(ns1:getMaterials) ::)) as element() (:: schema-element(ns2:GetMaterials) ::)  {
    <ns2:GetMaterials>
        <ns2:request>
            {
                if ($requestElement/ns1:emirateCode)
                then <ns2:EmirateCode>{fn:data($requestElement/ns1:emirateCode)}</ns2:EmirateCode>
                else ()
            }
            {
                if ($requestElement/ns1:materialCode)
                then <ns2:MaterialCode>{fn:data($requestElement/ns1:materialCode)}</ns2:MaterialCode>
                else ()
            }
            {
                if ($requestElement/ns1:materialDescAr)
                then <ns2:MaterialDescAr>{fn:data($requestElement/ns1:materialDescAr)}</ns2:MaterialDescAr>
                else ()
            }
            {
                if ($requestElement/ns1:materialDescEn)
                then <ns2:MaterialDescEn>{fn:data($requestElement/ns1:materialDescEn)}</ns2:MaterialDescEn>
                else ()
            }
            {
                if ($requestElement/ns1:userId)
                then <ns2:UserID>{fn:data($requestElement/ns1:userId)}</ns2:UserID>
                else ()
            }
            {
                if ($requestElement/ns1:systemCode)
                then <ns2:SystemCode>{fn:data($requestElement/ns1:systemCode)}</ns2:SystemCode>
                else ()
            }
        </ns2:request>
    </ns2:GetMaterials>
};

local:func($requestElement)
"""

output_xslt_file = 'C:\\Users\\inbra\\pyscripts\\xquery\\loutput.xsl'

xquery_to_xslt(xquery_code, output_xslt_file)


```
